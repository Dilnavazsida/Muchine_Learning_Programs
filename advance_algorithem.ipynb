```python
# Applying advance image filter

# 1) Bilateral (for Advance Smothing Effect)

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("rose.webp")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
bilateral = cv2.bilateralFilter(img,d=200,sigmaColor=50,sigmaSpace=50)

plt.subplot(1,2,1),plt.imshow(img),plt.title("source img"),plt.axis("off")
plt.subplot(1,2,2),plt.imshow(bilateral),plt.title("Smothe image using Bilateral img"),plt.axis("off")
```

```python
# 2) unsharp effect for (sharpping the image)
import matplotlib.pyplot as plt 
import cv2

img = cv2.imread("mountain.webp")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#  blur
gauss = cv2.GaussianBlur(img,(9,9),10)

# unsharp masking
unsharp = cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(1,2,2),plt.imshow(img),plt.title("Sharping image using  unsharp"),plt.axis("off")
```
