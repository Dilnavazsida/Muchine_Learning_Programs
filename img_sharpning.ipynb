```python
# image sharpning

import cv2
import matplotlib.pyplot as plt
import numpy as np 

img = cv2.imread("my_photo.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

k1 = np.array([[-1,-1,-1],[-1,10,-1],[-1,-1,-1]])
k2 = np.array([[1,1,1],[1,-7,1],[1,1,1]])
k4 = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

k1_img = cv2.filter2D(img,-1,k1)
k2_img = cv2.filter2D(img,-1,k2)
k4_img = cv2.filter2D(img,-1,k4)

plt.figure(figsize=(10,6))
plt.subplot(2,3,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(2,3,2),plt.imshow(k1_img),plt.title("k1_img img"),plt.axis("off")
plt.subplot(2,3,3),plt.imshow(k2_img),plt.title("k2_img img"),plt.axis("off")
plt.subplot(2,3,4),plt.imshow(k4_img),plt.title("k4_img img"),plt.axis("off")
```
