```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
k1 = np.ones((4,4),np.float32)/16
img1 = cv2.filter2D(img,-1,k1)
plt.subplot(1,2,1),plt.imshow(img),plt.title("orignal img"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(img1),plt.title("created img"),plt.axis('off')
```

```python
import matplotlib.pyplot
import cv2
import numpy as np

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kl = np.array([[-1,-1,-1],[-1,12,-1],[-1,-1,-1]])
kloutput = cv2.filter2D(img,-1,kl)
plt.subplot(1,2,1),plt.title('Source img'),plt.imshow(img)
plt.subplot(1,2,2),plt.title('original img'),plt.imshow(kloutput)
```
