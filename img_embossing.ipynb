```python
import matplotlib.pyplot as plt 
import cv2
import numpy as np

img =  cv2.imread('a2.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

k1 = np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
k2 = np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
k3 = np.array([[1,0,0],[0,0,0],[0,0,-1]])
k4 = np.array([[0,0,1],[0,0,0],[-1,0,0]])

img1 = cv2.filter2D(img,-1,k1)
img2 = cv2.filter2D(img,-1,k2)
img3 = cv2.filter2D(img,-1,k3)
img4 = cv2.filter2D(img,-1,k4)

plt.subplot(2,3,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(2,3,2),plt.imshow(img1),plt.title("k1 img"),plt.axis("off")
plt.subplot(2,3,3),plt.imshow(img2),plt.title("k2 img"),plt.axis("off")
plt.subplot(2,3,4),plt.imshow(img3),plt.title("k3 img"),plt.axis("off")
plt.subplot(2,3,5),plt.imshow(img4),plt.title("k4 img"),plt.axis("off")
```
