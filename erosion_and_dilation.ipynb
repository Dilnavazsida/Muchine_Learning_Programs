```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("j.png")
img = cv2.bitwise_not(img)

k1 = np.ones((3,3),np.uint8)

erosion = cv2.erode(img,k1,iterations = 1)

dilation = cv2.dilate(img,k1,iterations = 1)

plt.subplot(1,3,1),plt.imshow(img,cmap="gray"),plt.title("source img"),plt.axis("off")
plt.subplot(1,3,2),plt.imshow(erosion,cmap="gray"),plt.title("erosion img"),plt.axis("off")
plt.subplot(1,3,3),plt.imshow(dilation,cmap="gray"),plt.title("dilation img"),plt.axis("off")
```
