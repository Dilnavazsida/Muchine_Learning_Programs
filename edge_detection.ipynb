```python
# Edge Detaction

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("my_photo.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# SOBEL 
# Syntex => Sobel(img,ddepth,x,y,kernal_size)

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,10)

sobel_combined = cv2.magnitude(sobelx,sobely)
plt.figure(figsize=(10,6))

plt.subplot(1,4,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(1,4,2),plt.imshow(sobelx),plt.title("sobelx img"),plt.axis("off")
plt.subplot(1,4,3),plt.imshow(sobely),plt.title("sobely img"),plt.axis("off")
plt.subplot(1,4,4),plt.imshow(sobel_combined),plt.title("sobel_combined img"),plt.axis("off")
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("my_photo.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# gassianblur
gass = cv2.GaussianBlur(img,(3,3),0)

# canny   Syntax => canny(img,threshold1,threhold2)

canny = cv2.Canny(gass,threshold1=50,threshold2=150)

# ploting and img

plt.subplot(2,3,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(2,3,2),plt.imshow(gass),plt.title("Gassian Blur img"),plt.axis("off")
plt.subplot(2,3,3),plt.imshow(canny),plt.title("Canny img"),plt.axis("off")
```
