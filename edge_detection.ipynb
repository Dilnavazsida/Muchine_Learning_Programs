```python
# Edge Detaction

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("my_photo.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# SOBEL 
# Syntex => Sobel(img,ddepth,x,y,kernal_size)

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,10)

sobel_combined = cv2.magnitude(sobelx,sobely)
plt.figure(figsize=(10,6))

plt.subplot(1,4,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(1,4,2),plt.imshow(sobelx),plt.title("sobelx img"),plt.axis("off")
plt.subplot(1,4,3),plt.imshow(sobely),plt.title("sobely img"),plt.axis("off")
plt.subplot(1,4,4),plt.imshow(sobel_combined),plt.title("sobel_combined img"),plt.axis("off")
```
