```python
# applying diffrent type of blur effect

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

box = cv2.blur(img,(10,10))

plt.subplot(1,2,1),plt.imshow(img),plt.title("source img"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(box),plt.title("Box Blur img"),plt.axis('off')
```

```python
# gaussian Blur

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

gos = cv2.GaussianBlur(img,(-12,-12),10)


plt.subplot(1,2,1),plt.imshow(img),plt.title("source img"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(gos),plt.title("gaussian Blur img"),plt.axis('off')


```

```python
# Median Blur

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

median = cv2.medianBlur(img,5)  # we can take any positive odd number as number increse so effect increse

plt.subplot(1,2,1),plt.imshow(img),plt.title('Source_img'),plt.axis("off")
plt.subplot(1,2,2),plt.imshow(median),plt.title('Updated_img'),plt.axis("off")
```

```python
# All Blur Types 

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# BOX EFFECT
box = cv2.blur(img,(10,10)) # you can take any positive integer , number increse so effect incerese

# Gaussian Blur
gass = cv2.GaussianBlur(img,(-10,-10),10)  # we can take any odd number as number incere effect incerese

# Meddian Blur
med = cv2.medianBlur(img,1) # we can take any positive odd number 

plt.figure(figsize=(10,3))

plt.subplot(2,2,1),plt.imshow(img),plt.title("Source img"),plt.axis("off")
plt.subplot(2,2,2),plt.imshow(box),plt.title("Box Blur"),plt.axis("off")
plt.subplot(2,2,3),plt.imshow(gass),plt.title("Gaussian Blur"),plt.axis("off")
plt.subplot(2,2,4),plt.imshow(med),plt.title("Median Blur"),plt.axis("off")
```
