```python
# applying diffrent type of blur effect

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('my_photo.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

box = cv2.blur(img,(10,10))

plt.subplot(1,2,1),plt.imshow(img),plt.title("source img"),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(box),plt.title("Box Blur img"),plt.axis('off')
```
